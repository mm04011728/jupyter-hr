```python
import jieba
import jieba.analyse
import jieba.posseg as pseg  
import os  
import sys  
from sklearn import feature_extraction  
from sklearn.feature_extraction.text import TfidfTransformer  
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics.pairwise import pairwise_distances_argmin
import numpy as np
```

```python
file = open(r"data\2018-03-18\北京_大数据.csv",mode='r',encoding='utf-8')
data = file.readlines()
file.close()
```

```python
def filter_words(words):
    list = []
    for k,f in words:  
        if f == "n" or f == "l" or f == "l" or f == "eng" or f == "nz":
            list.append(k)
    return ' '.join(list)
    
```

```python
corpus = [filter_words(pseg.cut(line.split(',')[10])) for i,line in enumerate(data) if i!=0]
```

```python
for line in corpus:
    print(line)
```

```python
vectorizer=CountVectorizer()    #该类会将文本中的词语转换为词频矩阵，矩阵元素a[i][j] 表示j词在i类文本下的词频  
transformer=TfidfTransformer()  #该类会统计每个词语的tf-idf权值  
tf = vectorizer.fit_transform(corpus)
tfidf=transformer.fit_transform(tf)#第一个fit_transform是计算tf-idf，第二个fit_transform是将文本转为词频矩阵  
word=vectorizer.get_feature_names()#获取词袋模型中的所有词语  
weight=tfidf.toarray()#将tf-idf矩阵抽取出来，元素a[i][j]表示j词在i类文本中的tf-idf权重  
```

```python
def print_word_weight(word,weight):
    for i in range(len(weight)):#打印每类文本的tf-idf词语权重，第一个for遍历所有文本，第二个for便利某一类文本下的词语权重  
        print(u"-------这里输出第",i,u"类文本的词语tf-idf权重------")
        for j in range(len(word)):
            if weight[i][j] != 0 :
                print(word[j],weight[i][j])
```

```python
print_word_weight(word,weight)
```

```python
X = StandardScaler().fit_transform(weight)
```

```python
db = DBSCAN(eps=0.01, min_samples=2).fit(X)
labels = db.labels_
```

```python
k_means_n = 8
k_means = KMeans(init='random', n_clusters=k_means_n, n_init=10)
k_means.fit(X)
k_means_cluster_centers = np.sort(k_means.cluster_centers_, axis=0)
k_means_labels = pairwise_distances_argmin(X, k_means_cluster_centers)
```

```python
set_labels = set(labels)
print(set_labels)
n_clusters_ = len(set_labels)
print(n_clusters_)
```

```python
list_cluster_db = [[] for i in set_labels]
for i,l in enumerate(labels):
    for j in set_labels:
        if l == j:
            list_cluster_db[l].append(data[i+1])
```

```python
list_cluster_kmeans = [[] for i in range(k_means_n)]
for i,l in enumerate(k_means_labels):
    for j in range(k_means_n):
        if l == j:
            list_cluster_kmeans[l].append(data[i+1])
```

```python
def print_result_cluster(list_cluster):
    for i,cluster in enumerate(list_cluster):
        print("第{0}类公司————————\r\n".format(i+1))
        for line in cluster:
            print(i)
            print(line)
```

```python
print_result_cluster(list_cluster_db)
```
